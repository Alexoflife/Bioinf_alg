## 2. Analysis Pipeline

### a). Aligning with HISAT2

`hisat2-build <reference.fasta> <genome index>`
    
`hisat2 -x [path to HISAT2 index] -U [FASTQ file] | samtools sort > out.bam` for 16, 17, 18, 19 fq files


### b) Quantifying with featureCounts


`conda install gffread`

`gffread GCF_000146045.2_R64_genomic.gff -T -o GCF_000146045.2_R64_genomic.gtf`

`featureCounts -g gen_id -a GCF_000146045.2_R64_genomic.gtf -o all_data all.bam`  # 16, 17, 18, 19

`cat all_data | cut -f 1,7-10 > simple_counts_all.txt` 


### c) Find differentially expressed genes with Deseq2



`cat simple_counts_all.txt | R -f deseq2.r`

`cat norm-matrix-deseq2.txt | R -f draw-heatmap.r`
