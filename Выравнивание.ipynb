```python
def global_alignment(A, B, match_score=3, gap_score=2):
	n, m, d= len(A), len(B), -gap_score
	def S(A, B):
		if A == B:
			return match_score
		else:
			return -match_score
	F = [[0] * (m+1) for i in range(n+1)]
	for i in range(n+1):
		F[i][0] = d * i
	for j in range(m+1):
		F[0][j] = d * j
	for i in range(1, n+1):
		for j in range(1, m+1):
			Match = F[i-1][j-1] + S(A[i-1], B[j-1])   
			Delete = F[i-1][j] + d
			Insert = F[i][j-1] + d
			F[i][j] = max(Match, Delete, Insert)
	Align_A, Align_B = "", ""
	while i > 0 and j > 0:
		Score, ScoreDiag,  ScoreLeft, ScoreUp = F[i][j], F[i-1][j-1], F[i][j-1], F[i-1][j]
		if Score == ScoreDiag + S(A[i-1], B[j-1]):
			Align_A = A[i-1] + Align_A
			Align_B = B[j-1] + Align_B
			i -= 1
			j -= 1
		elif Score == ScoreUp + d:
			Align_A = A[i-1] + Align_A
			Align_B = "_" + Align_B
			i -= 1
		elif Score == ScoreLeft + d:
			Align_A = "_" + Align_A
			Align_B = B[j-1] + Align_B
			j -= 1
	while i > 0:
		Align_A = A[i-1] + Align_A
		Align_B = "_" + Align_B
		i -= 1
	while j > 0:
		Align_A = "_" + Align_A
		Align_B = B[j-1] + Align_B
		j -= 1
	return Align_A + "\n" + Align_B


def local_alignment(A, B, match_score=3, gap_score=2):
	n, m, d, maximum = len(A), len(B), -gap_score, 0
	def S(A, B):
		if A == B:
			return match_score
		else:
			return -match_score
	F = [[0] * (m+1) for i in range(n+1)]
	for i in range(1, n+1):
		for j in range(1, m+1):
			Match = F[i-1][j-1] + S(A[i-1], B[j-1])   
			Delete = F[i-1][j] + d
			Insert = F[i][j-1] + d
			F[i][j] = max(Match, Delete, Insert, 0)
			maximum = max(maximum, F[i][j])
			if maximum == F[i][j]:
				start_i = i
				start_j = j
	i, j = start_i, start_j
	Align_A, Align_B = "", ""
	if F[i][j] == 0:
		if m < n:
			Align_A = A
			Align_B = B + "_" * (n-m)
		elif n < m:
			Align_A = A +  "_" * (m-n)
			Align_B = B
		elif n == m:
			Align_A = A
			Align_B = B
	else:
		while i > 0 and j > 0:
			Score, ScoreDiag,  ScoreLeft, ScoreUp = F[i][j], F[i-1][j-1], F[i][j-1], F[i-1][j]
			if Score == 0:
				Align_A = A[i-1] + Align_A
				Align_B = B[j-1] + Align_B
				i -= 1
				j -= 1
			if Score == ScoreDiag + S(A[i-1], B[j-1]):
				Align_A = A[i-1] + Align_A
				Align_B = B[j-1] + Align_B
				i -= 1
				j -= 1
			elif Score == ScoreUp + d:
				Align_A = A[i-1] + Align_A
				Align_B = "_" + Align_B
				i -= 1
			elif Score == ScoreLeft + d:
				Align_A = "_" + Align_A
				Align_B = B[j-1] + Align_B
				j -= 1
		while i > 0:
			Align_A = A[i-1] + Align_A
			Align_B = "_" + Align_B
			i -= 1
		while j > 0:
			Align_A = "_" + Align_A
			Align_B = B[j-1] + Align_B
			j -= 1
		while start_i < n and start_j < m:
			Align_A += A[start_i]
			Align_B += B[start_j]
			start_i += 1
			start_j += 1
		while start_i < n:
			Align_A += A[start_i]
			Align_B += "_"
			start_i += 1
		while start_j < m:
			Align_A += "_"
			Align_B += B[start_j]
			start_j += 1
	return Align_A + "\n" + Align_B

print(global_alignment("fusdfhis", "sdfuifhs"), local_alignment("fusdfhis", "sdfuifhs"), sep = "\n\n")
```
